```python
from google import genai

# Блок 1: Чтение входных данных
# Загрузка содержимого 'article.txt' для последующего анализа
with open("article.txt", "r", encoding="utf-8") as file:
    long_article = file.read()

# Формирование промта и payload
# Использование Few-Shot Prompting для обеспечения строгого вывода в формате JSON
my_prompt = f"""
Ты - высококвалифицированный преподаватель, который составляет вопросы для теста

Инструкция:
1. Прочитай предоставленный ниже текст.
2. Сгенерируй ТРИ вопроса по содержанию текста.
3. Ответ должен быть представлен ИСКЛЮЧИТЕЛЬНО в формате JSON.

ПРИМЕР ВЫХОДНОГО JSON (для имитации):
{{
  "questions": [
    {{
      "id": 1,
      "topic": "История",
      "question": "Кто создал язык Python?",
      "answer_type": "text"
    }}
  ]
}}

Текст для анализа:
{long_article}
"""

# Запрос к модели (inference)
# Инициализация клиента и вызов Gemini 2.5 Flash для генерации контента
client = genai.Client(api_key=API_KEY)
response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=my_prompt
)

# Сохранение ответа в  summary.txt
# Сохранение JSON ответа модели в файл 'summary.txt'
if response.text:
    with open("summary.txt", "w", encoding="utf-8") as file:
        file.write(response.text)

print("✅ Скрипт успешно отработал и сохранил файл!")
```
