```python
from google import genai

# Load source data
with open("reviews.txt", "r", encoding="utf-8") as file:
    all_reviews = file.read()

# Prompt engineering with Few-Shot technique
my_prompt = f"""
Ты - аналитик службы поддержки. Твоя задача - разобрать отзывы клиентов.

Инструкция:
Для каждого отзыва определи:
1. Тональность (Позитивная, негативная или нейтральная).
2. Категорию (Доставка, качество еды, сервис или меню).

Ответ выдай строго в формате JSON.

Пример входа:
{{
  "reviews_analysis": [
    {{
      "id": 1,
      "sentiment": "Позитивная",
      "category": "Сервис",
      "summary": "Краткое описание сути"
    }}
  ]
}}

Отзывы для анализа:
{all_reviews}
"""


# API Client initialization and request
client = genai.Client(api_key=API_KEY)
response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=my_prompt
)



# Output processing and saving
if response.text:
    with open("analysis_results.txt", "w", encoding="utf-8") as file:
        file.write(response.text)
    print("✅ Анализ завершен! Проверь файл analysis_results.txt")
```
